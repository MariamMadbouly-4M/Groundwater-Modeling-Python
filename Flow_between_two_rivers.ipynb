```python
#confined Flow between two parallel rivers

import numpy as np
import matplotlib.pyplot as plt
#parameters
l=1000
H=10
zb=-6
k=10
n=0.3
T=k*H
h0=6
hl=4

#equations
x=np.linspace(0,l,100)   #create array of 100 numbers start from 0 and end by L
h=(hl-h0)*x/l+h0   #sloution for the head a straight line
Qx=-T *(hl-h0)/l * np.ones_like(x)   # discarge vector 

#basic plot
# Plot 1: Hydraulic Head (Potentiometric Surface)
plt.subplot(121)
plt.subplot(121)
plt.plot(x, h, 'b-', label='Hydraulic Head (h)')
plt.axhline(y=zb+H, color='k', linestyle='--', label='Top of Aquifer')
plt.axhline(y=zb, color='brown', label='Bottom of Aquifer (zb)')
plt.title('Confined Flow: Head')
plt.xlabel('Distance (m)')
plt.ylabel('Elevation (m)')
plt.legend()
plt.grid(True, alpha=0.3)
#plot
plt.figure(figsize=(12, 5))
# Plot 2: Specific Discharge (Qx)
plt.subplot(122)
plt.plot(x, Qx, 'r-')
plt.title('Discharge (Qx)')
plt.xlabel('Distance (m)')
plt.ylabel('Flow Rate ($m^2/d$)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
```
